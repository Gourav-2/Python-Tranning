```python
# Q3. Write a Python program that takes a string and a character as
# input and returns a list of indices where the character occurs in the
# string.
# Input: "programming", 'm'
# Output: [6, 7]

x=input("enter a string :")
y=input("enter char")
z=[]
for i in range(len(x)):
    if x[i]==y:
        z.append(i)

print(z)
        

```

```python
# Q4. Write a Python program to remove all occurrences of a given
# substring from a string.
# Input: "This is a test string for testing", "test"
# Output: "This is a string for ing"

x=input("enter a string :")
y=input("enter word")
z=[]
x=x.split(" ")
for i in x:
    if i!=y:
        z.append(i)

z=" ".join(z)
print(z)


```

```python
# 5 Write a Python program to find all unique characters in a given
# string.
# Input: "programming"
# Output: ['p', 'o', 'r', 'g', 'a', 'm', 'i', 'n'] -->

a=list(input("enter string"))

b=[]
for i in range(len(a)):
    if a[i]==a[len(a)-1]:
        b.append(a[i])
    elif a[i]!=a[i+1]:
        
        b.append(a[i])

print(b)

```
